```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import knn

file = "datingTestSet2.txt"
data, labels = knn.readData(file)
fig = plt.figure()
ax = fig.add_subplot(111)
plt.xlabel('Game time')
plt.ylabel('Ice cream')
ax.scatter(data[:,1],data[:,2])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.xlabel('Game time')
plt.ylabel('Ice cream')
ax.scatter(data[:,1],data[:,2],15.0*np.array(labels),15.0*np.array(labels))
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.xlabel('Fly miles')
plt.ylabel('Game time')
ax.scatter(data[:,0],data[:,1],15.0*np.array(labels),15.0*np.array(labels))
plt.show()
```
